### Obtención de enlaces con los datos

Esta función nos permite recolectar los enclaces de todos los fabricantes existentes.

```python
def get_marcas():
    src = requests.get("https://www.kimovil.com/es/todas-las-marcas-de-moviles")
    soup = BeautifulSoup(src.content, "lxml")
    marcas = soup.find("ul",{"class":"navigation-list", "id":"inline-search-list"}).find_all("li", {"class":"item"}, "lxml")
    res = []
    for marca in marcas:
        if marca.find("a") is not None:
            res.append(marca.find("a").get("href"))

    return res
```

Basandonos en los enclaces de la función anterior, permite buscar los enlaces de los moviles con un nombre coincidente en nuestra base de datos de dxomark.csv

```python
def get_moviles():
    enlaces = get_marcas()
    m = list(pd.read_csv("dxomark.csv",sep=";")["Device"])
    for i in range(len(m)):
        m[i]=m[i].lower().replace(" (exynos)","")
    moviles_disp = []
    moviles = []
    for link in enlaces:
        src = requests.get(link)
        soup = BeautifulSoup(src.content, "lxml")
        for elem in soup.find_all("ul", {"class":"simple-device-list"}):
            for phone in elem.find_all("li"):
                if phone.find("div",{"class":"title"}) is not None:
                    if phone.find("div",{"class":"title"}).text.strip().lower() in m:
                        moviles_disp.append(phone.find("a").get("href"))
                        moviles.append(phone.find("div",{"class":"title"}).text.strip().lower())
                
                #Si solo se desea buscar moviles con precio descomentar esto.
                """if phone.get("data-info") is not None: 
                    d = json.loads(phone.get("data-info"))
                    if d["countryPrices"] is not None:
                        if "es" in d["countryPrices"]:
                            if d["countryPrices"]["es"] != "":
                                moviles_disp.append(phone.find("a").get("href"))"""
                                
    return moviles_disp, moviles #devuelve una lista con los enlaces disponibles y los moviles encontrados
```

### Recolección de datos.

Despues de localizar los datos que nos interesan de la página web, diseñamos una función que los permita recolectar de forma automatizada.

```python
column_names = ["Nombre", "Marca", "Val_Ki", "Val_Usu", "Pantalla", "Cpu","Puntuacion_CPU","Resolucion_Pantalla",
                "Peso", "Bateria", "Precio","Sistema_Operativo", "Ram", "Tipo_RAM","Nanometros", "Tipo_pantalla","Numero_camaras", 
                "Tipo_camara_1", "Resolucion_1", "Sensor_1","Tipo_1", "Apertura_1", "Tamaño_pixel_1", "Tamaño_sensor_1", 
                "Tipo_camara_2","Resolucion_2", "Sensor_2","Tipo_2", "Apertura_2", "Tamaño_pixel_2", "Tamaño_sensor_2",
                "Tipo_camara_3","Resolucion_3", "Sensor_3","Tipo_3", "Apertura_3", "Tamaño_pixel_3", "Tamaño_sensor_3",
                "Estabilizacion_optica"]
```

```python
def get_attrs(url):
    attrs=dict(zip(column_names, [np.nan for i in range(len(column_names))] ))
    result = requests.get(url)
    web = result.content
    soup = BeautifulSoup(web)
    #MODELO Y MARCA
    name = soup.find("h1", id ="sec-start" ).text.replace("Precio y características del", "").replace("\n", "")
    marcaTable = soup.find("table", {"class": "k-dltable"})
    marca = marcaTable.find("td").text.replace("\n", "")
    #PRECIO
    precio1=soup.find("div", {"class": "container-datasheet"})
    precio2=precio1.find("section", {"class": "kc-container white container-sheet-intro"})
    precio3=precio2.find("div", {"class": "k-dl version-prices-table-wrap"})
    precio4=precio3.find("td" ,{"class":"js_kps"})
    if precio4.get("data-info") is not None: 
                    d = json.loads(precio4.get("data-info"))
                    if d["countryPrices"] is not None:
                        if "es" in d["countryPrices"]:
                            if d["countryPrices"]["es"] != "":
                                attrs["Precio"] = d["countryPrices"]["es"]
    #VALORACIONES
    valoracion_Ki=soup.find(class_="value").text.replace("\n", "")  #Si no tienen precio no devuelve valoracion
    valoracion_us = soup.find("div", {"class": "score"}).text.replace("\n", "")
    #SISTEMA OPERATIVO
    so=soup.find("ul", {"class": "kiui-grid k-main k-auto-column device-mini-datasheet device-mini-datasheet-sheet"})
    so2= so.find_all("div", {"class": "wrapp jump-to"})[-1]
    sistema_operativo = so2.find_all(("span", {"class": "spec sec max-3"}))[2].text
    #RAM
    ram1 = soup.find("div", {"class": "lay-main"})
    ram2 = ram1.find("section", {"class": "kc-container white container-sheet-hardware"})
    ram3 = ram2.find_all("table", {"class": "k-dltable"})[2]
    ram = ram3.find("td").text
    #TIPO RAM
    tipo_ram1 = soup.find("div", {"class": "lay-main"})
    tipo_ram2 = tipo_ram1.find("section", {"class": "kc-container white container-sheet-hardware"})
    tipo_ram3 = tipo_ram2.find_all("table", {"class": "k-dltable"})[2]
    tipo_ram = tipo_ram3.find_all("td")[1].text.replace("\n", "")
    #PUNTUACIÓN CPU
    punt_cpu1 = soup.find("div", {"class": "lay-main"})
    punt_cpu2 = punt_cpu1.find("section", {"class": "kc-container white container-sheet-hardware"})
    punt_cpu3 = punt_cpu2.find_all("table", {"class": "k-dltable"})[3]
    punt_cpu = punt_cpu3.find("div",{"class": "wrap"}).text.split()[0].replace(".","")
    #NANÓMETROS PROCESADOR
    nano1 = soup.find("div", {"class": "lay-main"})
    nano2 = nano1.find("section", {"class": "kc-container white container-sheet-hardware"})
    nano3 = nano2.find("table", {"class": "k-dltable"})
    nano = nano3.find_all("td")[3].text.replace("\n", "")
    #ESPECIFICACIONES
    #pantalla
    tableScreen = soup.find("li", {"class": "item item-screen k-rowspan-2"})
    screen = tableScreen.find("span", {"class": "spec main"}).text
    tecno_pantalla = tableScreen.find_all("span", {"class": "spec main"})[1].text
    resolucion_pantalla = tableScreen.find("span", {"class": "spec sec"}).text
    #peso
    tableWeight = soup.find("li", {"class": "item item-size k-rowspan-2"})
    weight = tableWeight.find("span", {"class": "spec main"}).text.replace("\n", "")
    #procesador
    cpu = soup.find("span", {"class":"spec main max-2"}).text.replace("\n", "")
    #battery
    battery = soup.find("li", {"class": "item item-battery"}).text.replace("\n", "")
    #CAMERAS
    camaras = soup.find("section", {"class":"kc-container dark black-isometric container-sheet-camera"}).find("div",{"class":"k-column-blocks g-2-cols"}).find_all("div", {"class":"w50"})
    attrs["Estabilizacion_optica"] = soup.find("section", {"class":"kc-container dark black-isometric container-sheet-camera"}).find("div", {"class":"wrapper"}).find("table",{"class":"k-dltable"}, recursive=False).find_all("td")[1].text.strip()
    attrs["Numero_camaras"]=len(camaras)
    for i, cam in enumerate(camaras):
        i=i+1
        attrs[f"Tipo_camara_{i}"]= cam.find("div",{"class":"k-h5"}).text.replace("\n","")
        for fila in cam.find_all("tr"):
            if fila.th.text.strip() == "Resolución":
                attrs[f"Resolucion_{i}"]=fila.td.text.replace("\n","").replace(" Mpx","").replace(" ","")
            if fila.th.text.strip() == "Sensor":
                attrs[f"Sensor_{i}"]=fila.td.text.replace("\n","")
            if fila.th.text.strip() == "Tipo":
                attrs[f"Tipo_{i}"]=fila.td.text.replace("\n","").replace(" ","")
            if fila.th.text.strip() == "Apertura":
                attrs[f"Apertura_{i}"]=fila.td.text.replace("\n","").replace(" ","").replace("ƒ/","")
            if fila.th.text.strip() == "Tamaño del pixel":
                attrs[f"Tamaño_pixel_{i}"]=fila.td.text.replace("\n","").replace(" ","").replace("µm","")
            if fila.th.text.strip() == "Tamaño del sensor":
                attrs[f"Tamaño_sensor_{i}"]=fila.td.text.replace("\n","").replace(" ","")
        for nombre, col in zip(cam.find_all("dt"),cam.find_all("dd")):
            if nombre.text.strip() == "Resolución":
                attrs[f"Resolucion_{i}"]= col.text.strip().replace("\n","").replace(" Mpx","").replace(" ","")
            if nombre.text.strip() == "Sensor":
                attrs[f"Sensor_{i}"]=col.text.strip().replace("\n","")
            if nombre.text.strip() == "Tipo":
                attrs[f"Tipo_{i}"]=col.text.strip().replace("\n","").replace(" ","")
            if nombre.text.strip() == "Apertura":
                attrs[f"Apertura_{i}"]=col.text.strip().replace("\n","").replace(" ","").replace("ƒ/","")
            if nombre.text.strip() == "Tamaño del pixel":
                attrs[f"Tamaño_pixel_{i}"]=col.text.strip().replace("\n","").replace(" ","").replace("µm","")
            if nombre.text.strip() == "Tamaño del sensor":
                attrs[f"Tamaño_sensor_{i}"]=col.text.strip().replace("\n","").replace(" ","")

    attrs["Nombre"] = name[1:]
    attrs["Marca"] = marca
    attrs["Val_Ki"] = valoracion_Ki
    attrs["Val_Usu"] = valoracion_us
    attrs["Pantalla"] = screen.replace('"',"")
    attrs["Peso"] = weight.replace("g", "")
    attrs["Cpu"] = cpu
    attrs["Bateria"] = battery.replace("mAh", "")
    attrs["Sistema_Operativo"] = sistema_operativo.replace("\n", " ")[1:-1]
    attrs["Ram"]=ram.replace("GB", "")
    attrs["Tipo_RAM"]=tipo_ram
    attrs["Nanometros"]=nano.replace("nm","")
    attrs["Puntuacion_CPU"]=punt_cpu
    attrs["Resolucion_Pantalla"]=resolucion_pantalla
    attrs["Tipo_pantalla"]= tecno_pantalla
    
    return attrs
```

Generamos un dataframe con todos los enlaces de la web.

```python
def get_datos(l):
    df = pd.DataFrame()
    for link in l:
        phone = get_attrs(link)
        df = df.append(phone, ignore_index = True, sort=False)
    return df

df = get_datos(enlaces + faltan)
df.head()
```

Guardamos una copia del dataframe 

```python
df = df[column_names]
df.to_csv("specs.csv", sep=";", index=False)
df.head
```

Transformamos los valores nulos, en verdaderos nulos y convertimos los "si" y "no" en booleanos.

```python
for columna in column_names:
    df[columna] = df[columna].replace('--', np.nan)
    df[columna] = df[columna].replace('Unknown', np.nan)
    df[columna] = df[columna].replace('Sí', "True")
    df[columna] = df[columna].replace('No', "False")
    
df.to_csv("specs.csv", sep=";", index=False)
```
